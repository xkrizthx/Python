```python
import matplotlib.pyplot as plt 
figura, xa=plt.subplots();
xa.plot([1,2,3,4],[1,2,4,0.5])  # Grafico de dispersion
#plt.savefig('diagramadispersion.png')
plt.show()
```

```python
import matplotlib.pyplot as plt 
figura, xa=plt.subplots();
xa.fill_between([1,2,3,4],[1,2,4,0.5])

```

```python
import matplotlib.pyplot as plt 
figura, xa=plt.subplots();
xa.bar([1,2,3,4],[1,2,4,0.5])
plt.show()
```
